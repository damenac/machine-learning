```python
def batch_gradient_descent(X,Y,W,alpha,iters):
    cost_history = [0] * iters
    #initalize our cost history list to store the cost function on every iteration
    for i in range(iters):
        prediction = np.dot(X,W.T)
        
        W = W - (alpha/len(Y)) * np.dot(prediction - Y,X)
        cost_history[i] = cost_function(X,Y,W)
    return W,cost_history


def cost_function(X,Y,W):
    m = len(Y)
    prediction = np.dot(X,W.T)
    return (1/(m)) * np.sum((prediction - Y) ** 2)
```

```python
epoch = 10000
batch_W, batch_history = batch_gradient_descent(X, Y, W, 0.1, epoch)
```
